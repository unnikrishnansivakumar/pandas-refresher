# Numpy and Pandas refresher

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points=np.arange(-5,5,0.01)
```

```python

x,y = np.meshgrid(points,points)
z   = np.sqrt(x**2 + y**2)
```

```python
plt.imshow(z,cmap=plt.cm.gray);
plt.show()
```

```python
#inverse

z=10-z
plt.imshow(z,cmap=plt.cm.gray);
plt.show()
```

```python
# absolute using where

d=np.where(z>5.9735,1,0)
plt.imshow(d,cmap=plt.cm.gray);
plt.show()
```

```python
#Check for number of blank pixels

1000*1000-np.sum(d)
```

```python
s=np.array(([1,0,1],[0,1,0],[1,0,1]))
plt.imshow(s,cmap=plt.cm.gray);
plt.show();
```

```python
#gives the x axis sum
sum_yaxis = np.sum(d,axis=1)
sum_xaxis = np.sum(d,axis=0)
```

```python
#Cumsum and cumprod 

f=s.cumsum(1)
plt.imshow(f,cmap=plt.cm.gray);
plt.show()
```

```python
# Can also sort by row and column

plt.imshow(np.sort(s,1),cmap=plt.cm.gray);
plt.show()
```

```python
# Function application and mapping

ex_frame=DataFrame(np.random.randn(4,3),index=['Delhi','Mumbai','Kolkata','Chennai'],columns=list('abc'))
ex_frame
```

```python
ex_frame.abs()
```

```python
print(ex_frame.sum(axis=0))
print(ex_frame.sum(axis=1))
```

```python
#Also incase of a defined function

somefunc=lambda x: max(x)-min(x)

ex_frame.apply(somefunc,axis=1)
```

```python
#if func returns multiple values then

def somefunc2(x):
    return Series([min(x),max(x),sum(x)],index=['min','max','sum'])

ex_frame.apply(somefunc2,axis=0)
```

```python
ex_frame.apply(somefunc2,axis=1)
```

```python
#Applying function elementwise using applymap

format = lambda x: '%.2f' %x
ex_frame.applymap(format)
```

```python
# Correlation and covariance

#importing pandas web data

import pandas.io.data as web

compdata={}

for ticker in ['MSFT','GOOG','AAPL','IBM']:
    compdata[ticker]=web.get_data_yahoo(ticker,'1/1/2014','1/1/2015')

```

```python
price = DataFrame({tic:data['Adj Close'] for tic,data in compdata.iteritems()})
volume = DataFrame({tic:data['Volume'] for tic,data in compdata.iteritems()})
```

```python
returns=price.pct_change()
```

```python
returns.cov()
```

```python
returns.corr()
```
